# Interactive Demo

This page contains **runnable code cells**. Click the rocket icon (🚀) in the
top-right toolbar and select **Live Code** to make them interactive.

A remote Binder kernel will be launched — all execution happens server-side.

```python
# This cell is tagged `thebe-init` — it runs automatically
# when the Thebe kernel starts, so imports are ready.
import numpy as np
import matplotlib.pyplot as plt
```

## Basic example

Try editing the code below and pressing **Shift + Enter** to re-run.

```python
x = np.linspace(0, 4 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(x, y, color="steelblue")
ax.set_title("Try changing the function!")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.tight_layout()
plt.show()
```

## Heavier computation

Because cells execute on a **remote server**, you can run operations that would
be too heavy for a static docs site:

```python
# Simulate a moderately expensive operation
from time import time

start = time()
big_matrix = np.random.randn(2000, 2000)
eigenvalues = np.linalg.eigvalsh(big_matrix @ big_matrix.T)
elapsed = time() - start

print(f"Computed {len(eigenvalues)} eigenvalues in {elapsed:.2f}s")
print(f"Top-5 eigenvalues: {eigenvalues[-5:][::-1].round(1)}")
```
